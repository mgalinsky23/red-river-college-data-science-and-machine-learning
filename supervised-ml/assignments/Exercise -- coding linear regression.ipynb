```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def absolute_trick(x, y, m, b, eta):
    ŷ = m * x + b
    if y >= ŷ:
        m = m + eta * x
        b = b + eta
    if y < ŷ:
        m = m - eta * x
        b = b - eta
        
    return m, b
```

```python
def square_trick(x, y, m, b, eta):
    ŷ = m * x + b
    m = m + eta * x * (y - ŷ)
    b = b + eta * (y - ŷ)
    
    return m, b
```

```python
def linear_regression(features, labels, eta, epochs = 10000):
    m = random.random()
    b = random.random()
    for epoch in range(epochs):
        i = random.randint(0, len(features) - 1)
        x = features[i]
        y = labels[i]
        m, b = square_trick(x, y, m, b, eta)

    return m, b
```

```python
def linear_regression_absolute(features, labels, eta, epochs = 10000):
    m = random.random()
    b = random.random()
    for epoch in range(epochs):
        i = random.randint(0, len(features) - 1)
        x = features[i]
        y = labels[i]
        m, b = absolute_trick(x, y, m, b, eta)

    return m, b
```

```python
features = np.array([1, 2, 3, 5, 6, 7])
labels = np.array([155, 197, 244, 356, 407, 448])

eta = 0.01
epochs = 10000
```

```python
m_square, b_square = linear_regression(features, labels, eta, epochs)
print(f'Slope (m) using square trick: {m_square}')
print(f'Y-intercept (b) using square trick: {b_square}')

plt.scatter(features, labels, color = 'red', label = 'Data Points')
line_x = np.arange(min(features), max(features) + 1)
line_y_square = m_square * line_x + b_square
plt.plot(line_x, line_y_square, label = 'Fitted Line (Square Trick)')
plt.xlabel('Features')
plt.ylabel('Labels')
plt.legend()
plt.show()
```

```python
m_absolute, b_absolute = linear_regression_absolute(features, labels, eta, epochs)
print(f'Slope (m) using absolute trick: {m_absolute}')
print(f'Y-intercept (b) using absolute trick: {b_absolute}')

plt.scatter(features, labels, color='red', label='Data Points')
line_y_absolute = m_absolute * line_x + b_absolute
plt.plot(line_x, line_y_absolute, label='Fitted Line (Absolute Trick)')
plt.xlabel('Features')
plt.ylabel('Labels')
plt.legend()
plt.show()
```

## Analysis of Model Fits

### Square Trick
- The slope obtained using the square trick is ~ 50.6152.
- The y-intercept obtained using the square trick is ~ 99.5051.
- Observing the plotted line, the square trick fits the data points well, closely following the trend of the data.

### Absolute Trick
- The slope obtained using the absolute trick is ~ 60.7388.
- The y-intercept obtained using the absolute trick is ~ 48.6003.
- In comparison, the absolute trick also aligns well with the trend of the data.

### Conclusion
- Both methods produce models that fit the data well. While the slopes and y-intercepts are slightly different, both lines appear to follow the overall trend of the data.
- There is no significant difference in the performance of the two methods. Both tricks can be used effectively for training linear regression models.
