# Load the dataset:

```python
import pandas as pd

# Load the dataset into a pandas DataFrame
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

# Explore the Dataset:

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(iris.head())  # Fill in the gap with .head() to display first 5 rows
print("\n")

# Print a concise summary of the dataset
print("Summary of the dataset:")
print(iris.info())  # Fill in the gap with .info() to print dataset info
print("\n")

# Output the dimensions of the dataset
print("Dimensions of the dataset:")
print(iris.shape)  # Fill in the gap with .shape to output dimensions
print("\n")

# Retrieve and print the names of all columns in the dataset
print("Columns in the dataset:")
print(iris.columns)  # Fill in the gap with .columns to print column names
print("\n")
```

# Adding a new column

```python
# Adding a new column 'PetalArea' based on petal dimensions
iris['PetalArea'] = iris['petal_length'] * iris['petal_width'] 
```

```python
# Adding a new column 'SepalArea' based on sepal dimensions
iris['SepalArea'] = iris['sepal_length'] * iris['sepal_width']
```

# Column Access

```python
# Column Access: Access the 'Species' column
species_column = iris[['species']]  # Fill in the gap to access 'Species' column

# Row Access: Access data of the iris with index 10
iris_index_10_data = iris.iloc[[10]]  # Fill in the gap for index 10

# Subset rows using loc: Access data of irises with SepalWidth greater than 3.5
irises_with_wide_sepal = iris.loc[iris['sepal_width'] > 3.5]  # Fill in the gap for SepalWidth condition

# Subset rows using iloc: Access the first 3 rows and first 2 columns of the dataset
first_3_rows_first_2_columns = iris.iloc[:3, :2]  # Fill in the gap to access first 3 rows and first 2 columns
```

# Conditional filtering

```python
# Display unique species names available in the dataset
print("Unique species in the dataset:")
print(iris['species'].unique())  # Fill in the gap to display unique species
print("\n")
```

```python
# Ask the user to choose a species and put .lower()
chosen_species = input("Enter a species to analyze (setosa, versicolor, virginica): ").strip().lower()

# Validate user input and filter data for the chosen species
if chosen_species in ['setosa', 'versicolor', 'virginica']:
    # Filter data for the chosen species
    chosen_species_data = iris[iris['species'] == chosen_species]

    # Display basic information about the chosen species data
    print(f"\nData for {chosen_species.capitalize()} species:")
    print(chosen_species_data.head())  # Fill in the gap to display the first few rows of chosen species data

    # Calculate mean sepal length for the chosen species
    mean_sepal_length = chosen_species_data['sepal_length'].mean()
    print(f"\nMean Sepal Length for {chosen_species.capitalize()} species: {mean_sepal_length:.2f}")  # Fill in the gaps for mean calculation
```

```python
# Example: Plotting histogram of sepal lengths for the chosen species
import matplotlib.pyplot as plt
if chosen_species in ['setosa', 'versicolor', 'virginica']:
    plt.figure(figsize=(8, 6))
    plt.hist(chosen_species_data['sepal_length'], bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Sepal Length Distribution for {chosen_species.capitalize()} Species')
    plt.xlabel('Sepal Length')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
    
else:
     print("Invalid species choice. Please choose from 'setosa', 'versicolor', or 'virginica'.")
```
