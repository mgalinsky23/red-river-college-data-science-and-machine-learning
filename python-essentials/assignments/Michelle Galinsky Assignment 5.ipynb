# Scenario 1: Analyzing Sales Data

You work for a retail company that sells electronics. Your task is to analyze the sales data for the past month to understand the performance of different products. This analysis will help in making decisions about inventory management and marketing strategies.



Dataset Description:

You have a dataset named sales_data.csv that includes the following columns:

Product: The name of the product sold.
Units Sold: The number of units sold for each product.
Price per Unit: The selling price of each unit.
Total Sales: Calculated as Units Sold * Price per Unit.
Tasks:

Calculate Basic Statistics:

Calculate the mean, median, mode, and range for the Units Sold column.

Compute the variance and standard deviation for the Total Sales column.

Quantile Analysis:

Determine the 25th, 50th (median), and 75th percentiles of the Price per Unit column.

Outlier Detection:

Identify outliers in the Total Sales column using the Interquartile Range (IQR) method. Any sales values that are 1.5 times the IQR above the third quartile or below the first quartile should be considered outliers.
Visualization:


Create a boxplot to visualize the distribution of Price per Unit across different products. This plot will help in identifying outliers visually and understanding the spread of prices

```python
# Imports:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats 
```

```python
data = pd.read_csv("C:/Users/miche/Downloads/dataset-week10/sample_sales_data.csv")
data['Total Sales'] = data['Units Sold'] * data['Price per Unit']
```

```python
# Calculate Basic Statistics:

# Calculate the mean, mode, and range for the Units Sold column

print('Mean:', np.mean(data['Units Sold']))
print('Mode:', data['Units Sold'].mode()[0])
print('Range:', data['Units Sold'].max() - data['Units Sold'].min())
      
# Compute the variance and standard deviation for the Total Sales column
print('Variance:', data['Total Sales'].var())
print('Standard Deviation:', data['Total Sales'].std())
```

```python
# Quantile Analysis:

# Determine the 25th, 50th (median), and 75th percentiles of the Price per Unit column
print('Quintiles (25th, 50th, 75th):', np.percentile(data['Price per Unit'], [25, 50, 75]))
```

```python
# Outlier Detection:

# Identify outlier in the Total Sales column using the IQR method. Any sales values 1.5 times the IQR above the 3rd quartile or below the 1st
# are outliers

#IQR = Q3 - Q1
Q1 = np.percentile(data['Total Sales'], 25)
Q3 = np.percentile(data['Total Sales'], 75)
IQR = Q3 - Q1
print('IQR:', IQR)

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = data[(data['Total Sales'] < lower) | (data['Total Sales'] > upper)]
print('Outliers:\n', outliers)
```

```python
# Visualization:

# Create a boxplot to visualize the distribution of Price per Unit across different products.
sns.boxplot(x='Product', y='Price per Unit', data=data)

# Add title and labels
plt.title('Price per Unit across different Products')
plt.xlabel('Price per Unit')
plt.ylabel('Products')

# Show plot
plt.show()
```

# Scenario 2: Examining Student Test Scores


Background:
You are a data scientist working for a school district. You have been tasked with analyzing the distribution of test scores among students to gain insights into their performance.

Tasks:

Generate normally distributed test score data.

Create a DataFrame to store the test scores.

Visualize the distribution of test scores using a histogram.

Calculate and print basic statistics: mean, median, mode, and standard deviation of the test scores.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set random seed for reproducibility
np.random.seed(42)

# Generate normally distributed test score data
mean_score = 75  # mean score
std_dev_score = 10  # standard deviation
num_students = 1000  # number of students

scores = np.random.normal(loc=mean_score, scale=std_dev_score, size=num_students)

# Create a DataFrame
data = pd.DataFrame(scores, columns=['Scores'])

# Plot the score distribution
sns.histplot(data['Scores'], kde=True, bins = 50, color = 'r')
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()

# Calculate basic statistics
print(f"Mean: {data['Scores'].mean():.2f} cm")
print(f"Median: {data['Scores'].median():.2f} cm")
print(f"Mode: {data['Scores'].mode()[0]:.2f} cm")
print(f"Standard Deviation: {data['Scores'].std():.2f} cm")
```
