# Scenario 1:
In a renowned medical research institute, a team of dedicated data scientists is working on an important project to analyze breast cancer patient data. Their goal is to uncover patterns and insights that could lead to better diagnosis, treatment, and follow-up care for patients. As part of this project, they have gathered data from various sources and need to integrate, manipulate, and visualize it to draw meaningful conclusions.

Characters:
    
Dr. Sarah Thompson - Lead Data Scientist

Dr. Michael Lee - Medical Oncologist

Emily Chen - Data Analyst

Alex Rodriguez - Data Visualization Expert

# Data Sources:
    
The team has collected three datasets:

1-Patient Information: This dataset contains demographic information about patients, including their unique PatientID, Name, Age, and Gender.
    
2-Medical Records: This dataset provides detailed information about the medical history of patients, including their unique RecordID, PatientID, Diagnosis (Benign or Malignant), and Treatment received (Surgery, Chemotherapy, Radiation).
    
3-Follow-Up Records: This dataset tracks the follow-up visits of patients, including their unique FollowUpID, PatientID, FollowUpDate, and Outcome (Recovered or Ongoing treatment).

# Task:
    
The team needs to perform several tasks to analyze the data and generate insights:

Data Integration: Merge the three datasets based on the PatientID to create a comprehensive dataset that includes all relevant information about each patient.
    
Data Filtering and Categorization: Filter out patients who are above 60 years old and categorize the remaining patients into three age groups: Young, Middle-aged, and Senior.
        
Risk Categorization: Develop a method to categorize patients based on their diagnosis and follow-up outcome into risk categories such as 'High Risk Recovered', 'High Risk Ongoing', and 'Low Risk'.
    
Count Patients in Age Groups: Use a while loop to count the number of patients in each age group until a total of 10 patients have been categorized.
    
Visualization: Create a pie chart to visualize the distribution of diagnoses (Benign vs. Malignant) among the patients.

```python
# 1. Loading the Datasets:
#Dr. Sarah starts by loading the datasets into their data analysis environment.

# Load the datasets
patients = pd.read_csv(r'C:\Users\miche\Downloads\patients.csv')
medical_records = pd.read_csv(r'C:\Users\miche\Downloads\medical_records.csv')
follow_up = pd.read_csv(r'C:\Users\miche\Downloads\follow_up.csv')
```

```python
#2. Joining the Datasets:
#Emily, the data analyst, merges the datasets to create a single master dataset.
# HINT: Use pd.merge() to join patients with medical_records and then with follow_up on 'PatientID' , inner join
df = pd.merge(patients, medical_records, on='PatientID')
df = pd.merge(df, follow_up, on='PatientID')
```

```python
#3. Data Manipulation:
#Dr. Michael suggests filtering out older patients to focus on those who are 60 or younger. Emily categorizes these patients into age groups
df = df[df['Age'] <= 60]
df['AgeGroup'] = df['Age'].apply(lambda x: 'Young' if x < 30 else ('Middle-aged' if x < 50 else 'Senior'))
```

```python
#4. Control Structures for Risk Categorization:
#Dr. Sarah designs a function to categorize patients based on their diagnosis and follow-up outcome.

# Function for categorization based on Diagnosis and Outcome
# HINT: Complete the function to categorize patients as 'High Risk Recovered', 'High Risk Ongoing', or 'Low Risk'
def categorize_patient(row):
    if row['Diagnosis'] == 'Malignant':
        if row['Outcome'] == 'Recovered':
            return 'High Risk Recovered'
        else:
            return 'High Risk Ongoing'
    else:
        return 'Low Risk'

df['RiskCategory'] = df.apply(categorize_patient, axis=1)
```

```python
#5. Counting Patients in Age Groups:
#Emily uses a while loop to count the number of patients in each age group until they have categorized 10 patients.
# While loop for counting patients
count = {'Young': 0, 'Middle-aged': 0, 'Senior': 0}
total_count = 0
index = 0

# HINT: Complete the while loop to count patients in each AgeGroup until total_count reaches 10
while total_count < 10:
    age_group = df.iloc[index]['AgeGroup']
    count[age_group] += 1
    total_count += 1
    index += 1

print("Age Group Counts:", count)
```

```python
# 6. Creating a Pie Chart:
#Alex, the data visualization expert, creates a pie chart to visualize the distribution of diagnoses
#Pie chart visualization
diagnosis_counts = df['Diagnosis'].value_counts()
diagnosis_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Diagnosis Distribution')
plt.ylabel('')
plt.show()
```

# Explain in your own words how the code worked for this scenario

**Explanation:**

First, we load the datasets into a DataFrame. Then, we merge multiple datasets using pd.merge() and specify the datasets and the column on which we want to merge the datasets. In this case, we first merge patients and medical_records to df on 'PatientID', then, df with follow_up on 'PatientID'.

After that, we filter out patients over 60 and assign the remaining into different age groups with the column name value 'AgeGroup'. We create a function to categorize patients based on if their dignosis is Malignant, and then based on that, if their follow up outcome is Recovered. This function then assigns three labels based on those two categories, which are then applied to a new column 'RiskCategory'.

Then, we create a loop to count the number of patients in each age group until the total count of individuals reaches 10.

Finally, we create a pie chart to visualize the data about the distribution of the diagnosis (Malignant vs. Benign).


# Scenario 2: Analyzing Wildfire Impact in Australia

In the Environmental Research Center, a team of data scientists is working on a critical project to analyze the impact of wildfires in Australia's jungles. The goal is to understand the patterns and effects of wildfires to improve prevention and response strategies. The team has gathered data from various sources, which need to be integrated, manipulated, and analyzed to draw meaningful conclusions.

# Characters:
    
Dr. Jane Foster - Lead Data Scientist

Dr. David Nguyen - Environmental Scientist

Sophie Martin - Data Analyst

Liam Patel - Data Visualization Expert

# Data Sources:

The team has collected three datasets:

Wildfire Incidents: This dataset contains information about wildfire incidents, including a unique IncidentID, Date, Location, and Severity (Low, Medium, High).

Weather Conditions: This dataset provides weather data, including a unique WeatherID, IncidentID, Temperature, Humidity, and Wind Speed.

Animal Sightings: This dataset tracks sightings of endangered animals in the affected areas, including a unique SightingID, IncidentID, AnimalType, and Count.

# Task:

The team needs to perform several tasks to analyze the data and generate insights:

Data Integration: Merge the datasets using a left join to include all wildfire incidents and their corresponding weather conditions and animal sightings.
Data Filtering and Categorization: Filter out incidents with low severity and categorize the remaining incidents based on temperature and wind speed.
Impact Analysis: Develop a method to categorize incidents based on the number of animal sightings into impact categories.
Visualization: Create a bar chart to visualize the number of incidents per severity category.

# 1. Loading the Datasets:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
wildfires = pd.read_csv(r'C:\Users\miche\Downloads\wildfires.csv')
weather = pd.read_csv(r'C:\Users\miche\Downloads\weather.csv')
sightings = pd.read_csv(r'C:\Users\miche\Downloads\sightings.csv')
```

# 2. Joining the Datasets:

```python
# Join the datasets
df = pd.merge(wildfires, weather, on='IncidentID', how='left')
df = pd.merge(df, sightings, on='IncidentID', how='left')
```

# 3. Data Manipulation:

```python
# Data manipulation
df = df[df['Severity'] != 'Low'] 
#Temperature column more than 30
df['TempCategory'] = df['Temperature'].apply(lambda x: 'Hot' if x > 30 else 'Mild')
#Wind Speed column more than 20
df['WindCategory'] = df['Wind Speed'].apply(lambda x: 'High Wind' if x > 20 else 'Low Wind')
```

# 4. Conditional Manipulation:

```python
# Function for impact categorization based on animal sightings
def categorize_impact(row):
    if row['Count'] > 50:
        return 'High Impact'
    elif row['Count'] > 20:
        return 'Moderate Impact'
    else:
        return 'Low Impact'

df['ImpactCategory'] = df.apply(categorize_impact, axis=1)
```

# 5. Visualization:

```python
# Bar chart visualization
severity_counts = df['Severity'].value_counts()
severity_counts.plot(kind='bar')
plt.title('Number of Incidents per Severity Category')
plt.xlabel('Severity')
plt.ylabel('Number of Incidents')
plt.show()
```

# Explain in your own words how the code worked for this scenario

**Explanation:**

First, we load the datasets into a DataFrame. Then, we merge multiple datasets using pd.merge() for wildfires and weather to df on 'IncidentID', then, df with sightings on 'IncidentID'. We specify the type of join (left join) by adding how = 'left' to the end of the merge statement.

After that, we filter out incidents where the severity is low, and create two new columns called 'TempCategory' and 'WindCategory' for the remaining data in which the temperature is greater than 50, and the wind speed is greater than 20, respectively. We create a function to categorize the impact based on animal sightings, and create three labels which are then added onto a new column called 'ImpactCategory'.

Finally, we create a bar graph to visualize the number of incidents per severity category.
