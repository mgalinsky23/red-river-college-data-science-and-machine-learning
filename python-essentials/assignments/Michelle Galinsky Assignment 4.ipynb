# Scenario: Analyzing Global Sales Performance

Background:
You are a data analyst at a global retail company named "GlobalMart." The company operates across multiple regions and sells a wide range of products. Management wants to understand the sales performance and trends over the years, the distribution of sales across different product categories, the relationship between sales and profit, and the top-performing regions. They have tasked you with analyzing the data and presenting your findings through visualizations.

Dataset:
The dataset global_sales_data.csv contains the following columns:

Year: The year of the sales data.

Region: The region where the sales occurred.

Product_Category: The category of the product sold.

Sales: The total sales amount in USD.

Units_Sold: The total number of units sold.

Profit: The total profit in USD.

# Sales Trend Over the Years (Line Plot):


```python
#Create a line plot to show the trend of total sales over the years for each region.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sales Trend Over the Years
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Sales', hue='Region', marker='o')
plt.title('Total Sales per Year per Region')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Regions:')
plt.show()
```

# Interpretation of Line Chart

```python
# Interpretation:

# Each line represents a region and it's total sales over the years. For example, the orange line is for Europe 
# (as seen in the legend). This line shows that this particular region had very good total sales till about 2018 
# where their sales started to drop. Meanwhile, the blue line (for North America) shows that this region had been 
# steadily dropping their total sales over the years.

# Next, I believe there are some outliers and unusual things going on in this chart. One of these is the sales trend 
# for the region of Africa. While the other regions have solid lines running for all the years of this dataset, Africa
# only has one value/point for the year of 2017. Another observation one could make is the sudden drop in the regions of 
# Asia and Europe. While this wouldn't be considered an outlier, I believe that it is unusual enough to mention in this analysis.
```

# Sales Share by Product Category (Pie Chart)

```python
# Filter data for the year 2020
df_2020 = df[df['Year'] == 2020]

# Aggregate sales by product category
sales_by_category = df_2020.groupby('Product_Category')['Sales'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
sales_by_category.plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Sales by Product categories')
plt.ylabel('')  # Hide the y-label
plt.show()
```

# Interpretation of Pie Chart

```python
# Interpretation:

# This pie chart illustrates the sales percentages of each product category for the year 2020. For instance, Clothing 
# accounted for 43.4% of the total sales in 2020. Another category, Furniture, shows the lowest sales at 26.4% for the 
# same year. This indicates a potential investment or improvement needed in this category by the "GlobalMart" company.
```

# Profit vs. Sales (Scatter Plot)

```python
# Profit vs. Sales Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Profit', y='Sales', hue='Region', style='Region', s=100)
plt.title('Profit vs. Sales')
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.show()
```

# Interpretation of Scatter Chart

```python
# Interpretation:

# This scatter plot shows the profit vs. sale for each region. The legend specifies each shape and the region
# it represents. This chart suggests a positive linear relationship between the profit and the sales. As the 
# profit increases, the sales increase, and vice versa. 

# There also appear to be 2 outliers. The regions of Africa and Europe lie on opposite sides of this chart. 
# Africa has the lowest profit vs. sales, while Europe lies at the average and the top parts of the plot.
```

# Top Performing Regions (Bar Chart)

```python
# Aggregate total sales by region
total_sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(5)

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=total_sales_by_region.index, y=total_sales_by_region.values, palette='viridis')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

# Interpretation of Scatter Chart

```python
# Interpretation:

# This bar chart displays the total sales by region, with Europe leading at 500000, and Asia following closely at just above 400000. 
# These are led by North and South America with 300000 and slightly over 200000, respectively. Lastly, Africa shows the lowest performance, 
# with sales of just under 100000.

# This chart displays a similar pattern to the previous ones, where Africa consistently underperforms, while Europe excels.
# This suggests that the company "GlobalMart" should invest in Africa while maintaining its strategies in Europe.
```

# 1. Swarm Plot: Age Distribution by Heart Disease

Use Seaborn to create a swarm plot to visualize the distribution of ages (age) between patients with and without heart disease (target).


Instructions:


Create a swarm plot where the x-axis represents the target (0 = no heart disease, 1 = heart disease) and the y-axis represents age.

Label the plot appropriately with a title, x-axis label, and y-axis label.

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Visualize the distribution of ages between patients with and without heart disease
import seaborn as sns
import matplotlib.pyplot as plt

sns.swarmplot(x='target', y='age', hue='target', data=heart_attack_data)
plt.xlabel('Heart Disease')
plt.ylabel('Age')
plt.title('Heart Disease by Age')
plt.show()
```

# Interpretation

```python
# Interpretation:

# This swarm plot illustrates the distribution of ages among patients with and without heart disease. The legend indicates what 
# the colors represent (0 for no heart disease and 1 for heart disease). The x-axis categorizes the groups, while the y-axis spans 
# patient ages, ranging from 30 to 80.

# From the chart, we observe a cluster of points around age 30 for patients without heart disease. As age increases, the number of 
# patients in this group decreases. In contrast, the group with heart disease (True) shows a rise in numbers as age progresses, especially 
# between 60 and 80 years.
```

# 2.Histogram Plot: Distribution of Maximum Heart Rate

Utilize Matplotlib to plot a histogram showing the distribution of maximum heart rate (thalach) among patients.

Instructions:

Create a histogram of maximum heart rate (thalach).

Customize the plot with appropriate bins, labels, and title.

```python
# Show the distribution of maximum heart rate among patients
import seaborn as sns
import matplotlib.pyplot as plt

plt.hist(heart_attack_data['thalach'], bins=25, edgecolor='black')
plt.title('Histogram of Maximum Heart Rate in Patients')
plt.xlabel('Heart Rate')
plt.ylabel('Number of Patients')

plt.show()
```

# Interpretation

```python
# Interpretation:

# This histogram shows the distribution of maximum heart rate among patients. It shows us the heart rate levels and the number of 
# patients at each level. In particular, we can see a spike in the number of patients who have heart rate levels of about 110, 135, and 195.
# Meanwhile, there are significant differences in the number of patients with levels of around 75, 115, and 155. 
# These indicate that there is a broad span of patients included in this dataset.
```

# 3.Regression Plot: Resting Blood Pressure vs. Serum Cholesterol

Create a regression plot using Seaborn to analyze the relationship between resting blood pressure (trestbps) and serum cholesterol levels (chol).

Instructions:

Generate a regression plot where trestbps is on the x-axis and chol is on the y-axis.

Interpret the slope and direction of the relationship based on the plot.

```python
# Analyze the relationship between resting BP and serum cholesterol levels
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x='trestbps', y='chol', data=heart_attack_data)
plt.title('Resting BP vs. Cholesterol Levels')
plt.xlabel('BPS')
plt.ylabel('Choelsterol')
plt.show()
```

# Interpretation

```python
# Interpretation:

# This regression plot illustrates the relationship between resting blood pressure (BPS) and cholesterol levels in patients. Each blue 
# dot represents a patient, while the solid blue line shows a general upward trend, indicating that higher resting blood pressure tends 
# to align with increased cholesterol levels.
```
